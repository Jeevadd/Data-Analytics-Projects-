```python
import pandas as pd
from textblob import TextBlob

# Load the CSV file
reviews = pd.read_excel(r"C:\Users\Jeeva D\OneDrive\Desktop\Forage Intern\British Airways\Book1.xlsx")

# Function to determine sentiment
def get_sentiment(review):
    analysis = TextBlob(review)
    # Classify the polarity: Positive, Negative, or Neutral
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Apply the function to the 'review' column
reviews['Sentiment'] = reviews['review '].apply(get_sentiment)

# Display the dataframe with the sentiment analysis
print(reviews)
reviews.to_csv(r"C:\Users\Jeeva D\OneDrive\Desktop\Forage Intern\British Airways\web dataset_sentime .csv" , index =False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
#data = pd.read_csv(r"C:\Users\Jeeva D\OneDrive\Desktop\Forage Intern\British Airways\web dataset_sentime.csv")
data = pd.read_csv(r"C:\Users\Jeeva D\OneDrive\Desktop\Forage Intern\British Airways\web dataset_sentime .csv")

# Filter the data for reviews that mention "service", "seat", "crew", or "staff"
keywords = ["service", "seat", "crew", "staff"]
filtered_data = data[data['review '].str.contains('|'.join(keywords), case=False, na=False)]

# Count the frequency of each unique sentiment in the filtered data
frequency = filtered_data['Sentiment'].value_counts()

# Create a bar plot using Matplotlib
plt.figure(figsize=(10, 6))
sns.barplot(x=frequency.index, y=frequency.values, palette='viridis')
plt.title('Frequency Chart of Sentiments for Reviews Mentioning "Service", "Seat", "Crew", or "Staff"')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
